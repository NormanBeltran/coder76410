```python
from openai import OpenAI
import os
from dotenv  import load_dotenv   # pip install python-dotenv
load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

contexto = "Eres un meteorologo experto, con reconocimientos mundiales en el cuidado medioambiental"
prompt = "¿Puedes darme información sobre el cambio climático?"

conversation = [
{ "role": "system", "content": contexto},
{ "role": "user", "content": prompt },
]

response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = conversation,
    max_tokens = 400,
    temperature = 0.9 # A mayor cercania de 1 mas creativa sera la respuesta que me dara el modelo
)

message = response.choices[0].message
print(message.role)
print(message.content)
```

```python
prompt2 = "¿Cómo podemos mitigar los Gases de efecto invernadero?"

conversation = [
    { "role": "system", "content": contexto},
    { "role": "user", "content": f"{message.role}: {message.content}" },
    { "role": "user", "content": prompt2 },
]

response2 = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = conversation,
    max_tokens = 400,
    temperature = 0.9
)

message = response2.choices[0].message
print(message.role)
print(message.content)
```
